### 3. Cho biết thuộc tính nào là thuộc tính phân loại, thuộc tính nào là thuộc tính định lượng

```python
# Your code here
import pandas as pd

# Đọc dữ liệu từ tệp CSV
data = pd.read_csv('/content/sample_data/car-sales.csv')

# Tạo danh sách các thuộc tính phân loại và định lượng
categorical_attributes = []
numeric_attributes = []

# Xác định thuộc tính phân loại và định lượng
for column in data.columns:
    if data[column].dtype == "object":
        categorical_attributes.append(column)
    else:
        numeric_attributes.append(column)

# In danh sách thuộc tính
print("Thuộc tính phân loại:")
for cat_attr in categorical_attributes:
    print(cat_attr)

print("\nThuộc tính định lượng:")
for num_attr in numeric_attributes:
    print(num_attr)
```

### 10. Khảo sát xe của hãng Honda

- Tính tỷ lệ xe Honda trong dữ liệu của cửa hàng (0.3)
- Cho biết giá trung bình của xe hãng Honda của cửa hàng (6500)
- Cho biết các xe của hãng Honda trong cửa hàng có những màu sắc nào (Red, Blue)

```python
# Your code here
import pandas as pd

# Đọc dữ liệu từ tệp CSV
data = pd.read_csv('/content/sample_data/car-sales.csv')

# 1. Tính tỷ lệ xe Honda trong dữ liệu của cửa hàng
ty_le_honda = len(data[data["Manufacturer"] == "Honda"]) / len(data)
print(f"Tỷ lệ xe Honda trong dữ liệu của cửa hàng: {ty_le_honda:.1%}")

# 2. Tính giá trung bình của xe hãng Honda của cửa hàng
gia_trung_binh_honda = data[data["Manufacturer"] == "Honda"]["Price"].mean()
print(f"Giá trung bình của xe hãng Honda của cửa hàng: ${gia_trung_binh_honda:.2f}")

# 3. Liệt kê các màu sắc của xe Honda trong cửa hàng
mau_sac_honda = data[data["Manufacturer"] == "Honda"]["Colour"].unique()
print("Các màu sắc của xe Honda trong cửa hàng:", ", ".join(mau_sac_honda))
```

# 11. Tính toán các giá trị sau của  thuộc tính giá ('Price')

- Giá trung bình các xe (7645)
- Giá trị trung vị các xe (6625)
- Miền giá trị của xe (18500)
- Miền phân vị của xe (2750)
- Điểm phân vị 0.4 và 0.6 của giá ([5750, 7000])

```python
# Your code here
import pandas as pd

# Đọc dữ liệu từ tệp CSV
data = pd.read_csv('/content/sample_data/car-sales.csv')

# Tính giá trung bình của xe
gia_trung_binh = data["Price"].mean()
print(f"Giá trung bình của xe: ${gia_trung_binh:.2f}")

# Tính giá trị trung vị của xe
gia_tri_trung_vi = data["Price"].median()
print(f"Giá trị trung vị của xe: ${gia_tri_trung_vi:.2f}")

# Tính miền giá trị của xe
gia_min = data["Price"].min()
gia_max = data["Price"].max()
mien_gia_tri = gia_max - gia_min
print(f"Miền giá trị của xe: ${mien_gia_tri}")


# Tính miền phân vị của xe (IQR)
q1 = data["Price"].quantile(0.25)
q3 = data["Price"].quantile(0.75)
iqr = q3 - q1
print(f"Miền phân vị của xe (IQR): ${iqr}")

# Tính điểm phân vị 0.4 và 0.6 của giá
percentile_40 = data["Price"].quantile(0.4)
percentile_60 = data["Price"].quantile(0.6)
print(f"Điểm phân vị 0.4 và 0.6 của giá: [${percentile_40}, ${percentile_60}]")

```
