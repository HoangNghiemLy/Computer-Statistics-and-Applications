### Bài 1: Tạo quần thể mô phỏng

Bạn hãy tạo một mẫu gồm 10000 phần tử mô phỏng chiều cao của nam thanh niên có giá trị từ 120cm - 200cm. Bạn hãy lưu kết quả vào biến **POP**.

Tính:
a. Chiều cao trung bình (kỳ vọng) của quần thể và độ lệch chuẩn về chiều cao của quần thể.

b. Tính tỷ lệ người cao trong quần thể, biết rằng thanh niên có chiều cao từ 180 trở lên được gọi là cao.

c. Vẽ histogram về chiều cao của quần thể. Theo bạn quần thể có phân phối chuẩn hay không?

d. Thử vẽ histogram và đồ thị hàm mật độ của phân phối chuẩn sử dụng tham số loc và scale bằng với kỳ vọng và độ lệch chuẩn của quần thể.

```python
import numpy as np
import matplotlib.pyplot as plt

# a. Tạo mẫu mô phỏng
np.random.seed(0)  # Để đảm bảo kết quả tái sản xuất
POP = np.random.uniform(120, 200, 10000)

# a. Chiều cao trung bình và độ lệch chuẩn
mean_height = np.mean(POP)
std_deviation = np.std(POP)

print("a. Chiều cao trung bình:", mean_height)
print("   Độ lệch chuẩn về chiều cao:", std_deviation)

# b. Tính tỷ lệ người cao (chiều cao >= 180)
tall_percentage = np.sum(POP >= 180) / len(POP)

print("b. Tỷ lệ người cao trong quần thể:", tall_percentage)

# c. Vẽ histogram chiều cao của quần thể
plt.hist(POP, bins=30, density=True, color='blue', alpha=0.7, label='Quần thể')
plt.xlabel("Chiều cao")
plt.ylabel("Tần số tương đối")
plt.title("Histogram chiều cao của quần thể")
plt.legend()

# Kiểm tra xem quần thể có phân phối chuẩn không
# d. Vẽ histogram và đồ thị hàm mật độ của phân phối chuẩn
from scipy.stats import norm

mu, sigma = mean_height, std_deviation

# Vẽ histogram của quần thể
plt.figure()
plt.hist(POP, bins=30, density=True, color='blue', alpha=0.7, label='Quần thể')
plt.xlabel("Chiều cao")
plt.ylabel("Tần số tương đối")

# Vẽ đồ thị hàm mật độ của phân phối chuẩn
x = np.linspace(120, 200, 100)
pdf = norm.pdf(x, mu, sigma)
plt.plot(x, pdf, 'r-', lw=2, label='Phân phối chuẩn')
plt.title("Histogram và đồ thị hàm mật độ của phân phối chuẩn")
plt.legend()

plt.show()
```

### Bài 2: Lấy mẫu cỡ 20

Bạn hãy lấy một mẫu kích thước 20 phần từ. Tính:

a. Trung bình mẫu, và độ lệch chuẩn của mẫu

b. Tỷ lệ người cao của mẫu

c. Vẽ histogram của mẫu

d. Thử vẽ đồ thị hàm mật độ của phân phối chuẩn với tham số loc và scale bằng với giá trị trung bình và độ lệch chuẩn và chiều cao của mẫu. Bạn có nhận xét gì không?

```python
# a. Lấy mẫu kích thước 20 từ quần thể
sample = np.random.choice(POP, size=20, replace=False)

# a. Trung bình mẫu và độ lệch chuẩn của mẫu
sample_mean = np.mean(sample)
sample_std_deviation = np.std(sample)

print("a. Trung bình mẫu:", sample_mean)
print("   Độ lệch chuẩn của mẫu:", sample_std_deviation)

# b. Tính tỷ lệ người cao trong mẫu
tall_percentage_sample = np.sum(sample >= 180) / len(sample)

print("b. Tỷ lệ người cao trong mẫu:", tall_percentage_sample)

# c. Vẽ histogram của mẫu
plt.hist(sample, bins=10, density=True, color='green', alpha=0.7, label='Mẫu')
plt.xlabel("Chiều cao")
plt.ylabel("Tần số tương đối")
plt.title("Histogram của mẫu")

# d. Thử vẽ đồ thị hàm mật độ của phân phối chuẩn với tham số loc và scale bằng với trung bình và độ lệch chuẩn của mẫu
x_sample = np.linspace(120, 200, 100)
pdf_sample = norm.pdf(x_sample, sample_mean, sample_std_deviation)
plt.plot(x_sample, pdf_sample, 'r-', lw=2, label='Phân phối chuẩn (Mẫu)')
plt.title("Histogram và đồ thị hàm mật độ của phân phối chuẩn (Mẫu)")
plt.legend()

plt.show()
```

### Bài 4: Khảo sát phân phối của trung bình mẫu

a. Thực hiện bài 2 100 lần, mỗi lần bạn tính được trung bình mẫu. Vẽ đồ thị histogram cho 100 trung bình mẫu bạn tính được.

b. Thực hiện bài 3 100 lần, mỗi lần bạn tính được trung bình mẫu. Vẽ đồ thị histogram cho 100 trung bình mẫu bạn tính được.

c. Vẽ histogram của hai câu a, b trong cùng một hình để so sánh. Theo bạn hình dạng của histogram thay đổi nói lên điều gì?

Gợi ý: tạo một danh sách rỗng để chứa các trung bình mẫu tính được, sử dụng hàm append() để thêm giá trị trung bình vào danh sách sau mỗi lần tính.

```python
# a. Thực hiện bài 2 100 lần và tính trung bình mẫu sau mỗi lần
sample_means_a = []

for _ in range(100):
    sample_a = np.random.choice(POP, size=20, replace=False)
    sample_mean_a = np.mean(sample_a)
    sample_means_a.append(sample_mean_a)

# Vẽ histogram cho 100 trung bình mẫu
plt.hist(sample_means_a, bins=20, density=True, color='blue', alpha=0.7, label='100 Trung bình mẫu (Bài 2)')
```

```python
# b. Thực hiện bài 3 100 lần và tính trung bình mẫu sau mỗi lần
sample_means_b = []

for _ in range(100):
    sample_b = np.random.choice(POP, size=100, replace=False)
    sample_mean_b = np.mean(sample_b)
    sample_means_b.append(sample_mean_b)

# Vẽ histogram cho 100 trung bình mẫu
plt.hist(sample_means_b, bins=20, density=True, color='green', alpha=0.7, label='100 Trung bình mẫu (Bài 3)')
```

```python
# c. Vẽ histogram của hai trường hợp
# Vẽ histogram của cả hai trường hợp trong cùng một hình
plt.hist(sample_means_a, bins=20, density=True, color='blue', alpha=0.7, label='100 Trung bình mẫu (Bài 2)')
plt.hist(sample_means_b, bins=20, density=True, color='green', alpha=0.7, label='100 Trung bình mẫu (Bài 3)')

plt.xlabel("Trung bình mẫu")
plt.ylabel("Tần số tương đối")
plt.title("Histogram của Trung bình Mẫu (Bài 2 và Bài 3)")
plt.legend()
plt.show()

```

### Bài 5: Khảo sát tỷ lệ mẫu

a. Bạn hãy mô phỏng 1000 lần lấy mẫu có kích cỡ 100. Vẽ histogram tỷ lệ người cao của các mẫu

b. Bạn hãy mô phỏng 1000 lần lấy mẫu có kích cỡ 500. Vẽ histogram tỷ lệ người cao của các mẫu

```python
# a. Mô phỏng 1000 lần lấy mẫu có kích thước 100
sample_sizes_a = 100
n_simulations_a = 1000
tall_percentages_a = []

for _ in range(n_simulations_a):
    sample_a = np.random.choice(POP, size=sample_sizes_a, replace=False)
    tall_percentage_a = np.sum(sample_a >= 180) / sample_sizes_a
    tall_percentages_a.append(tall_percentage_a)

# Vẽ histogram tỷ lệ người cao
plt.hist(tall_percentages_a, bins=20, density=True, color='blue', alpha=0.7, label='1000 Tỷ lệ mẫu (kích thước 100)')
```

### Bài 6: Khảo sát phương sai mẫu

Bạn hãy mô phỏng 1000 lần lấy mẫu có kích cỡ 100. Vẽ histogram đại lượng phương sai của các mẫu (Phương sai bằng bình phương độ lệch chuẩn)

```python
# Mô phỏng 1000 lần lấy mẫu có kích thước 100
sample_sizes = 100
n_simulations = 1000
sample_variances = []

for _ in range(n_simulations):
    sample = np.random.choice(POP, size=sample_sizes, replace=False)
    sample_variance = np.var(sample)  # Phương sai
    sample_variances.append(sample_variance)

# Vẽ histogram đại lượng phương sai
plt.hist(sample_variances, bins=20, density=True, color='blue', alpha=0.7, label='1000 Phương sai mẫu (kích thước 100)')
plt.xlabel("Phương sai")
plt.ylabel("Tần số tương đối")
plt.title("Histogram của Đại lượng Phương sai Mẫu (kích thước 100)")
plt.legend()
plt.show()
```

### Bài 7:  Bài Tập Ước Lượng

Khảo sát cân nặng (kg) của gà khi xuất chuồng, người ta cân một số con và có kết quả như sau:

> 2.1, 1.8, 2.0, 2.3, 1.7, 1.5, 2.0, 2.2, 1.8

Giả sử cân nặng của gà là biến ngẫu nhiên có phân phối chuẩn. Hãy lập khoảng tin cậy cho cân nặng trung bình của  khi xuất chuồng với độ tin cậy là 95%. Trong 2 trường hợp:

a, Biết $\sigma = 0.3$

b, Không biết $\sigma$


```python
import numpy as np

data = [2.1, 1.8, 2.0, 2.3, 1.7, 1.5, 2.0, 2.2, 1.8]

sample_mean = np.mean(data)
sample_std_deviation = np.std(data, ddof=1)  # Sử dụng ddof=1 để tính độ lệch chuẩn của mẫu (s)

confidence_level = 0.95
alpha = 1 - confidence_level

# Số mẫu
n = len(data)

# Độ lệch chuẩn của quần thể đã biết (σ)
population_std_deviation = 0.3

# Tính giá trị critical từ phân phối chuẩn (z-score)
z_critical = abs(norm.ppf(alpha / 2))  # Sử dụng hàm norm.ppf của SciPy

# Tính khoảng tin cậy
margin_of_error = z_critical * (population_std_deviation / np.sqrt(n))
confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)

print("a. Khoảng tin cậy 95% cho cân nặng trung bình (σ=0.3):", confidence_interval)
```

```python
from scipy.stats import t

confidence_level = 0.95
alpha = 1 - confidence_level

# Độ lệch chuẩn của mẫu đã tính ở phần a
sample_std_error = sample_std_deviation / np.sqrt(n)

# Số tự do (degrees of freedom) cho phân phối Student's t
degrees_of_freedom = n - 1

# Tính giá trị critical từ phân phối Student's t
t_critical = abs(t.ppf(alpha / 2, df=degrees_of_freedom))

# Tính khoảng tin cậy
margin_of_error_t = t_critical * sample_std_error
confidence_interval_t = (sample_mean - margin_of_error_t, sample_mean + margin_of_error_t)

print("b. Khoảng tin cậy 95% cho cân nặng trung bình (Không biết σ):", confidence_interval_t)
```
