### Câu 1:

```python
import matplotlib.pyplot as plt 
import numpy as np
x = np.linspace(0.0,4.0,5)
plt.title('First Graph!')
plt.plot(x,x**2,'r--',label ="$y=x^2$")
plt.plot(x,2*x,'b--',label = 'y=2x',marker = '^',mfc = 'blue')
plt.plot([2],[4],marker = '^',mfc ='blue')
plt.legend()
plt.xlabel('x- axis')
plt.ylabel('y - axis')
plt.show()



```

### Câu 2:

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0.0,10.0,10)
plt.subplot(1,2,1)
plt.plot(x,2*x,'b--',marker='.',mfc='blue',label = '$y=2*x$')
plt.legend()
plt.subplot(1,2,2)
plt.plot(x,x**2,'g-',marker='*',mfc = 'green',label = '$y=x^2$')
plt.legend()
plt.show()
```

### Câu 3:

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0,10,10)
plt.title('title')
plt.plot(x,x**2,color = 'blue',marker = 'o',mfc = 'red',label = 'y = x**2')
plt.legend()
plt.plot(x,x**3,'orange',marker = '*',mec = 'black',mfc = 'black',label = 'y = x**3')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
