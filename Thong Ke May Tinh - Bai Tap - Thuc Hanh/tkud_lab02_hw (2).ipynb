### Câu 2:

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0.0,10.0,9)
plt.subplot(1,2,1)
plt.plot(x,2*x,'b--',marker='o',ms=3,label='y = 2 * x')
plt.legend()
plt.subplot(1,2,2)
plt.plot(x,x**2,marker='*',color = 'g',label='y = x^2')
plt.legend()
```

```python
import matplotlib.pyplot as plt
import numpy as np
barwidth = 0.35

last_week_cups = (20, 35, 30, 35, 27)
this_week_cups = (25, 32, 34, 20, 25)
names = ['Mary', 'Paul', 'Billy', 'Franka', 'Stephan']
plt.title("Coffee Consumption")
plt.ylabel("Numbers of cups")
br1 = np.arange(len(last_week_cups))
br2 = [x + barwidth for x in br1]
last_week_cups1 = np.array([20, 35, 30, 35, 27])
this_week_cups1 = np.array([25,32,34,20,25])
plt.bar(br1, last_week_cups1,width = barwidth,label = "Last week")
plt.bar(br2 , this_week_cups,width = barwidth,label = "This week")
plt.legend()
plt.xticks([r + 0.17 for r in range(len(last_week_cups))],['Mary', 'Paul', 'Billy', 'Franka', 'Stephan'])
plt.show()
```
