```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0,4,5)
plt.plot(x,2*x,'b--',marker = '^',label='y=2x')
x = np.linspace(0,2.55,10)

plt.plot(x,x**2,color = 'red',label='$y=X^2$')
plt.legend()
x1 = np.linspace(2.55,4,10)
plt.plot(x1,x1**2,'b--',color = 'r')
plt.legend()
plt.show()

```

### Câu 2:

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0.0,10.0,10)
plt.subplot(1,2,1)
plt.plot(x,2*x,'b--',marker='.',mfc='blue',label = '$y=2*x$')
plt.legend()
plt.subplot(1,2,2)
plt.plot(x,x**2,'g-',marker='*',mfc = 'green',label = '$y=x^2$')
plt.legend()
plt.show()
```

```python
x = np.linspace(0, 100, 100)
y = x + np.random.randn(len(x))*5
z = x + np.random.randn(len(x))*20
```

```python
# Your code here
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 100, 100)
y = x + np.random.randn(len(x))*5
z = x + np.random.randn(len(x))*20

plt.title("Mô phỏng đồ thị y = x với các sai số khác nhau")
plt.xlabel("x")
plt.ylabel("y")
plt.plot(x,x,color='red')
plt.scatter(x,y,color='MAGENTA',label = 'random error in [0, 5]')
plt.scatter(x,z,color='orange',label = 'random error in [0, 20]')
plt.legend()
plt.show()
```
