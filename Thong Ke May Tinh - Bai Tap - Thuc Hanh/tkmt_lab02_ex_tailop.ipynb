### Câu 4: Định dạng đường thẳng

Phát sinh x là các điểm dữ liệu từ $[0, \pi]$. Bạn hãy vẽ đồ thị $y = sin(x)$ và $y = cos(x)$ như dưới đây.
Gợi ý: sử dụng hằng số pi và hàm sau:
- `np.pi` : số $\pi$
- `np.linspace()`
- `np.sin()`
- `np.cos()`

```python
import numpy as np
import matplotlib.pyplot as plt
x=np.linspace(0,2*np.pi,100)

plt.plot(x,np.sin(x),label = "sin(x)")
plt.plot(x,np.cos(x),'b--',color = 'orange',label = "cos(x)")
plt.grid(True)
plt.legend()
plt.title('Đồ thị hàm sin(x) và cos(x)')
plt.show
```

### Câu 5: Vẽ các đồ thị con

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0,2*np.pi,100)
fig = plt.figure()
fig.suptitle("Đồ thị hàm sin(x) và cos(x)")
plt.subplot(2,1,1)
plt.plot(x,np.sin(x),label ="sin(x)",color = 'b')
plt.legend()
plt.subplot(2,1,2)
plt.plot(x,np.cos(x),label = "cos(x)",color= 'r')
plt.legend()

```
